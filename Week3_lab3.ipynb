```python
# Task 1
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('images/test_image.jpg')
print(image.shape)
# 1-> translation_matrix = np.float32([[1, 0, 200], [0, 1, 100]])
translation_matrix = np.float32([[1, 0, -80], [0, 1, -60]])
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))
print(translated_image.shape)
plt.imshow(cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB))
plt.title("Translated image")
plt.axis("off")
plt.show()
```

```python
# Task 2
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('images/test_image.jpg')
print(image.shape)

(h, w) = image.shape[:2]
center = (w // 2, h // 2)
# rotation_matrix = cv2.getRotationMatrix2D(center, 90, 1.0)
rotation_matrix = cv2.getRotationMatrix2D(center, 180, 1.0)

rotated_image = cv2.warpAffine(image, rotation_matrix, (w, h))

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
# plt.title('Rotated Image (90 degrees)')
plt.title('Rotated Image (180 degrees)')
plt.axis('off')
plt.show()

```

```python
# Task 3
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('images/test_image.jpg')
if image is None:
    print("Image not found")
else:
    h, w = image.shape[:2]

    Factor = 1.5
    scaling_matrix = cv2.getRotationMatrix2D((0, 0), 0, Factor)

    scaled_image = cv2.warpAffine(image, scaling_matrix, (int(w*Factor), int(h*Factor)))

    plt.imshow(cv2.cvtColor(scaled_image, cv2.COLOR_BGR2RGB))
    plt.title("Scaled Image (1.5x)")
    plt.axis("off")
    plt.show()

    print("New picture:", scaled_image.shape)

```
